```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
import joblib
```

```python
# Importation de la base de donn√©e 
digits = load_digits()

x = digits.data
y = digits.target

print(f'dimension de X: {x.shape}')
```

```python
# Visualisation d'un de ces chiffres
plt.imshow(digits['images'][0], cmap='Greys_r')

# Entrainement du modele
model = KNeighborsClassifier()
model.fit(x, y)
model.score(x, y)
```

```python
# Tester le model
test = digits['images'][100].reshape(1, -1)
plt.imshow(digits['images'][100], cmap='Greys_r')
model.predict(test)
```

```python
# Tester le model
test = digits['images'][9].reshape(1, -1)
plt.imshow(digits['images'][9], cmap='Greys_r')
model.predict(test)
```

```python
import random
nombre = random.randint(0, 1797)
# Tester le model
test = digits['images'][nombre].reshape(1, -1)
plt.imshow(digits['images'][nombre], cmap='Greys_r')
model.predict(test)
```

```python
# Sauvegardez le mod√®le dans un fichier
joblib.dump(model, "modele_knn.pkl")
```

```python
def trained_model(model, donnee):
    from sklearn.neighbors import KNeighborsClassifier
    import joblib

    # Chargez le mod√®le
    loaded_model = joblib.load(model)

    # Nouvelles donn√©es d'entr√©e
    nouvelles_donnees = donnee

    # Faites des pr√©dictions
    predictions = loaded_model.predict(nouvelles_donnees)

    # Affichez les pr√©dictions
    print("Pr√©dictions :", predictions)
```

```python
nombre = random.randint(0, 1797)
# Tester le model
test = digits['images'][nombre].reshape(1, -1)
plt.imshow(digits['images'][nombre], cmap='Greys_r')

trained_model('modele_knn.pkl', test)
```
